```python
# Dependencies
import pandas as pd
import requests
import json
from config import api_key
from pprint import pprint
```

```python
youtube_data_df = pd.read_csv("youtube_video_data.csv")
cat_id_list = youtube_data_df["category_id"].unique()
```

```python
cat_id=list(cat_id_list)

query_url = f"https://www.googleapis.com/youtube/v3/videoCategories?part=snippet&regionCode=US&key={api_key}"
response = requests.get(query_url).json()

pprint(response)
```

```python
id_list=[]
cat_name=[]

item = response['items']

for i in range(len(item)):
    id_list.append(item[i]['id'])
    cat_name.append(item[i]['snippet']['title'])
```

```python
youtube_categories_df = pd.DataFrame({"id": id_list, 
                          "category":cat_name})
youtube_categories_df
```

```python
# Export data to CSV
youtube_categories_df.to_csv("output/YouTube_Categories.csv")
```
