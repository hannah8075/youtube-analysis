```python
videos['channel_id'] = ""

for index, row in videos.iterrows():
    vidid = row['video_id']
    video_url = f"https://www.googleapis.com/youtube/v3/videos?part=snippet&id={vidid}&key={api_key}"
    response = requests.get(video_url).json()
    
    
    
    try: 
   
        videos.loc[index,'channel_id'] = response['items'][0]['snippet']['channelId']
    
    except:
        
         pass

    

videos
```

```python
videos["subscriber_count"] = ""
videos["view_count"] = ""
videos["video_count"] = ""

for index, row in videos.iterrows():
    channel_title = row["channel_id"]
    query_url = f"https://www.googleapis.com/youtube/v3/channels?part=statistics&id={channel_title}&key={api_key2}"
    response = requests.get(query_url).json()

    try: 
        videos.loc[index,"subscriber_count"] = response["items"][0]["statistics"]["subscriberCount"]
        videos.loc[index,"video_count"] = response["items"][0]["statistics"]["videoCount"]
        videos.loc[index,"view_count"] = response["items"][0]["statistics"]["viewCount"]
    
    except:
        
        pass
    

videos
```
