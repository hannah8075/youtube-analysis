```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
```

```python
videos_csvpath = os.path.join("output/videos_data.csv")
videos_df = pd.read_csv(videos_csvpath)
youtube_csvpath = os.path.join("output/youtube2018.csv")
youtube_df = pd.read_csv(youtube_csvpath)
youtube_df
```

```python
merged_df = pd.merge(youtube_df, videos_df, on = "channel_title")
merged_df.groupby(["subscriber_count"])
merged_df
```

```python
sub_bins = [0, 499999, 999999, 1499999, 1999999, 2499999, 2999999, 3499999, 
            3999999, 4499999, 4999999, 5499999, 5999999, 6499999, 6999999, 7499999, 
            7999999, 8499999, 8999999, 9499999, 9999999, 80000000]

bin_titles = ["0-499999", "500000-999999", "1000000-1499999", "1500000-1999999", "2000000-2499999", "2500000-2999999", 
              "3000000-3499999", "3500000-3999999", "4000000-4499999", 
              "4500000-4999999", "5000000-5499999", "5500000-5999999", 
              "6000000-6499999", "6500000-6999999", "7000000-7499999", 
              "7500000-7999999", "8000000-8499999", "8500000-8999999", "9000000-9499999", "9500000-9999999", "10000000+"]

merged_df["subscriber range"] = pd.cut(merged_df["subscriber_count"], sub_bins, labels=bin_titles, include_lowest=True)

merged_df.head()
```

```python
subscribers_df = merged_df.groupby(["subscriber range"])
subscribers_df.describe()
```

```python
subscribers_data = subscribers_df["subscriber range"].value_counts()
subscribers_data

plt.bar(bin_titles, subscribers_data.values)

plt.xticks(rotation=90)
plt.ylabel("Number of Videos")
plt.xlabel("Number of Subscribers")
plt.title("Subscriber Count")
plt.show()
```
