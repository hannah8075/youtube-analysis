```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
```

```python
csvpath = os.path.join("output/videos_data.csv")
video_df = pd.read_csv(csvpath)
video_df
```

```python
subscribers = video_df["subscriber_count"].value_counts()
plt.bar(subscribers.index.values, subscribers.values)
plt.xticks(rotation=90)
plt.ylabel("Subscriber Count")
plt.xlabel("Number of Videos")
plt.show()
```
