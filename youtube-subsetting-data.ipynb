```python
# Preview data
youtube_df = pd.read_csv("youtube_video_data.csv")
youtube_df.head()
```

```python
# Create new columns
youtube_df["publish_dow"] = ""
youtube_df["publish_date"] = ""
youtube_df["publish_mhs"] = ""
youtube_df["trending year"] = ""
```

```python
# Decompose publish_time column

def DayOfWeek(time):
    #time format should be '2017-11-13'
    return datetime.datetime.strptime(time[:10], "%Y-%m-%d").strftime('%a')

def PublishDate(time): 
    return datetime.datetime.strptime(time[:10], "%Y-%m-%d").strftime('%m/%d/%Y')

def PublishTime(time):
    return time[11:19]


for index, row in youtube_df.iterrows():      
    # Get date at each row
    publish_time = row["publish_time"]
    youtube_df.loc[index, "publish_dow"] = DayOfWeek(publish_time)
    youtube_df.loc[index, "publish_date"]= PublishDate(publish_time)
    youtube_df.loc[index, "publish_mhs"]= PublishTime(publish_time)

youtube_df.head()
```

```python
youtube_df.count()
```

```python
# Change trending date format 
def convert(time): 
    return datetime.datetime.strptime(time, '%y.%d.%m').strftime('%m/%d/%y')

def TrendYear(time): 
    return datetime.datetime.strptime(time, '%y.%d.%m').strftime('%Y')

for index, row in youtube_df.iterrows():      
    # Get date at each row
    trend_date = row["trending_date"]
    #trend_date format %y.%d.%m
    formatted_date = convert(trend_date)
    #formatted date = %m/%d/%y
    youtube_df.loc[index, "trending year"]= TrendYear(trend_date)
    youtube_df.loc[index, "trending_date"]= formatted_date  
```
